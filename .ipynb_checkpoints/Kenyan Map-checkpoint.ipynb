```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap

fig = plt.figure(figsize=(10, 10))
m = Basemap(llcrnrlat=-5, urcrnrlat=5,
            llcrnrlon=33, urcrnrlon=43,
            lat_0=1, lon_0=50,)
m.drawcountries()
m.drawcoastlines()
#m.drawrivers()
m.drawstates()
m.bluemarble()
m.drawmapboundary(fill_color='aqua')
m.fillcontinents(color='#cc9955', lake_color='aqua')

# Map (long, lat) to (x, y) for plotting
x, y = m(37.9062, 0.0236)
plt.plot(x, y, 'ok', markersize=8)
plt.text(x, y, ' Nairobi', fontsize=18)
plt.savefig('kenya.png')
plt.savefig('kenya.pdf')
```
